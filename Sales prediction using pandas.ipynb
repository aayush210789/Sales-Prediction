```python
columns = ['Store', 'DayOfWeek', 'Promo']
```

```python
medians = train.groupby( columns )['Sales'].median()
```

```python
medians = medians.reset_index()
```

```python
test2 = pd.merge( test, medians, on = columns, how = 'left' )
```

```python
assert( len( test2 ) == len( test ))
```
